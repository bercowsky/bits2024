# Bivariate Analysis

Import libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

Read data

```python
df = pd.read_csv("../data/interim/with_all_indicators.csv")
```

## 1. Explanatory-Target

1. Sex-Target

```python
# Group the data by 'target' and 'Sex' to calculate raw counts
grouped = df.groupby(['target', 'Sex']).size().unstack(fill_value=0)

# Reset index and melt the dataframe for plotting
grouped_melted = grouped.reset_index().melt(id_vars='target', var_name='Sex', value_name='count')

# Create the bar plot
ax = sns.barplot(data=grouped_melted, x='target', y='count', hue='Sex')

# Calculate percentages within each 'target' group and add them as annotations
for container in ax.containers:
    for bar in container:
        height = bar.get_height()
        # Calculate the total count for each target group
        total_count = sum([b.get_height() for b in container])  # Total count for each target group
        percentage = (height / total_count) * 100  # Percentage within the target group
        ax.annotate(f'{percentage:.1f}%', 
                    (bar.get_x() + bar.get_width() / 2, height), 
                    ha='center', va='bottom', fontsize=10)

# Add labels and title
plt.xlabel('Target')
plt.ylabel('Count')
plt.title('Count Distribution of Sex by Target (Percentages on Bars)')

# Display the plot
plt.show()
```
